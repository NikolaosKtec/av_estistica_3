```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from functions import media_producao_total,plot_prod_total,producao_total_estados,var_por_estados
data = pd.read_excel('BDAgro.xlsx')
# print(data)
pd.__version__
%matplotlib inline
```

<p>(-qual a produção total e a media por UF, -qual a produção total
por tipo de cultivo, - qual a temperatura, umidade e precipitação total e media por UF)</p>

```python
RN, PB, CE = var_por_estados(data)
prod_rnT, prod_pbT, prod_ceT  = producao_total_estados(RN,PB,CE)
plot_prod_total(prod_rnT, prod_pbT, prod_ceT)
```

```python
mean_rnT,mean_pbT,mean_ceT = media_producao_total(RN,PB,CE)
fig, ax = plt.subplots(figsize=(4,4))

estados = ['RN', 'PB', 'CE']
counts = [mean_rnT, mean_pbT, mean_ceT]
bar_labels = ['RN', 'PB', 'CE']
bar_colors = ['tab:gray', 'tab:blue', 'tab:red',]

ax.bar(estados, counts, label=bar_labels, color=bar_colors,width=[0.5])

ax.set_ylabel('media de produção total')
ax.set_title('media de produção total')
ax.legend(title='Estados')
# ax.xaxis.offsetText.set_fontsize('x-small')

plt.show()
```

```python
prod_total_bi =  np.sum(RN['PBi']) + np.sum(PB['PBi']) + np.sum(CE['PBi'])
prod_total_bd =  np.sum(RN['PBd'])+ np.sum(PB['PBd']) + np.sum(CE['PBd'])

fig, ax = plt.subplots(figsize=(4,4))

tipos_batata = ['Batata inglesa', 'Batata doce']
counts = [prod_total_bi, prod_total_bd]
bar_labels = ['Batata inglesa', 'Batata doce']
bar_colors = ['tab:gray', 'tab:blue']

ax.bar(tipos_batata, counts, label=bar_labels, color=bar_colors,width=[0.5])

ax.set_ylabel('prod total por tipo')
ax.set_title('prod total por tipo')
ax.legend(title='batata')
# ax.xaxis.offsetText.set_fontsize('x-small')

plt.show()
```

```python
temp_me_rn =  np.mean(RN['temp'])
tem_me_pb =  np.mean(PB['temp'])
tem_me_ce =  np.mean(CE['temp'])

um_me_rn = np.mean(RN['umidade']) 
um_me_pb = np.mean(PB['umidade'])
um_me_ce = np.mean(CE['umidade'])

prec_me_rn = np.mean(RN['prec'])
prec_me_pb = np.mean(PB['prec'])
prec_me_ce = np.mean(CE['prec'])

fig, ax = plt.subplots(figsize=(12,4))

leg_x = ['temp_me_rn','temp_me_pb','temp_me_ce','um_me_rn','um_me_pb','um_me_ce','prec_me_rn','prec_me_pb','prec_me_ce']
counts = [temp_me_rn, tem_me_pb,tem_me_ce,um_me_rn,um_me_pb,um_me_ce,prec_me_rn,prec_me_pb,prec_me_ce]
bar_labels =  ['temp_me_rn','temp_me_pb','temp_me_ce','um_me_rn','um_me_pb','um_me_ce','prec_me_rn','prec_me_pb','prec_me_ce']
bar_colors = ['tab:gray', 'tab:blue','tab:pink', 'tab:gray', 'tab:blue','tab:pink','tab:gray', 'tab:blue','tab:pink',]

ax.bar(leg_x, counts, label=bar_labels, color=bar_colors,width=[0.5])

ax.set_ylabel('temperatura, umidade, precipitação medias')
ax.set_title('temperatura, umidade, precipitação medias')
ax.legend(title='batata')
# ax.xaxis.offsetText.set_fontsize('x-small')

plt.show()
```
